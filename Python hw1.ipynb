```python
import numpy as np #用np替代numpy

k=8700 #k是履約價
Premium_Call=254 #每一紙選擇權合約的價格
Premium_Put=268

Interval=800
ST=np.arange(k-Interval,k+Interval) #7900,9500
#ST=到期時標的物價格

y1=Profit_LongCall=np.maximum(ST-k,0)-Premium_Call
y2=Profit_ShortCall=-Profit_LongCall

import matplotlib.pyplot as plt #matplotlib是pyplot子庫提供繪圖API，用來繪製2D圖表

fig, ax = plt.subplots()
ax.spines['top'].set_visible(False) # 移除上方界線
ax.spines['right'].set_visible(False) # 移除下方界線
ax.spines['bottom'].set_position('zero') # 把x軸設定在中間
ax.tick_params(top=False, right=False) # 移除上方&右方tick-marks

plt.plot(ST,y1)
plt.plot(ST,y2)
plt.xlabel('ST') #x軸代表ST
plt.ylabel('Payoff/loss') #y軸代表Payoff
plt.title('+C&-C') #圖表標題

plt.plot(ST,y1,lw=1.5,label='Long Call',color = "yellow") #lw表示線寬
plt.plot(ST,y2,lw=1.5,label='Short Call',color = "magenta" )


plt.legend(loc=0) #執行跳出左上角框框(搭配多條線的label)，位置0代表最佳位置
plt.grid(True) #加底虛線
plt.axis('tight') #調整大小
plt.show()
```

```python
import numpy as np 
#用np替代numpy
k1=8700 #k是履約價
k2=8900
Premium_LongCall=350 #每一紙選擇權合約的價格
Premium_ShortCall=275

Interval=1200
ST=np.arange(k1-Interval,k1+Interval)
#ST是到期時標的物價值

y1=Profit_LongCall=np.where(ST > k1,((ST - k1) - Premium_LongCall) , -Premium_LongCall )
y2=Profit_ShortCall=np.where(ST > k2,((k2 - ST) + Premium_ShortCall) , Premium_ShortCall )
y3=Profit_Bullspread=y1+y2

import matplotlib.pyplot as plt #matplotlib是pyplot子庫提供繪圖API，用來繪製2D圖表

fig, ax = plt.subplots()
ax.spines['top'].set_visible(False) # 移除上方界線
ax.spines['right'].set_visible(False) # 移除下方界線
ax.spines['bottom'].set_position('zero') # 把x軸設定在中間
ax.tick_params(top=False, right=False) # 移除上方&右方tick-marks


plt.xlabel('ST') #x軸代表ST
plt.ylabel('Payoff') #y軸代表Payoff
plt.title('+C(k1)&-C(k2)') #圖表標題

plt.plot(ST,y1,lw=1.5, label='Long Call',color = "yellow",linestyle ='--') #lw表示線寬
plt.plot(ST,y2,lw=1.5,label='Short Call',color = "magenta",linestyle ='--')
plt.plot(ST,y3,lw=3,label='Bullspread',color = "red")


plt.legend(loc=0) #執行跳出左上角框框，位置0代表最佳位置
plt.grid(True) #加底虛線
plt.axis('tight')
plt.show()
```
